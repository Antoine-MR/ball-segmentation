```python
from pathlib import Path
import numpy as np
from PIL import Image
import os
```

```python
def load_mask_points(txt_path: Path):
    line = txt_path.read_text().strip()
    vals = list(map(float, line.split()))
    coords = vals[1:]
    return list(zip(coords[0::2], coords[1::2])) 


def colorize_pixels(img_path: Path, txt_path: Path, color=(0, 255, 0)):
    img = Image.open(img_path).convert("RGB")
    w, h = img.size
    points_norm = load_mask_points(txt_path)

    xs = (np.array([p[0] for p in points_norm]) * w).astype(int)
    ys = (np.array([p[1] for p in points_norm]) * h).astype(int)

    arr = np.array(img)
    xs = np.clip(xs, 0, w - 1)
    ys = np.clip(ys, 0, h - 1)

    arr[ys, xs] = color 
    return Image.fromarray(arr), img
```

```python
img_index = 420
TXT_PATH = Path("txt_output_folder") /os.listdir("txt_output_folder")[img_index]
IMG_PATH = Path("balls") / os.listdir("balls")[img_index]
```

```python
if IMG_PATH.exists() and TXT_PATH.exists():
    img_colored, img_original = colorize_pixels(IMG_PATH, TXT_PATH, color=(0, 255, 0))
    display(img_colored)
else:
    print(f"IMG_PATH exists: {IMG_PATH.exists()}")
    print(f"TXT_PATH exists: {TXT_PATH.exists()}")
    print("Please set valid IMG_PATH and TXT_PATH to test.")
```
