```python
# Imports des modules locaux
from pathlib import Path
import os
import time
from tqdm import tqdm

# Imports depuis le package src/
from src.pipeline import img_pipeline
from src.detection import RoboflowDetector
from src.segmentation import SAMSegmenter
```

```python
# Configuration
ROBOFLOW_API_KEY = "KDf14v839AjRzJXanN4h"
ROBOFLOW_MODEL_ID = "raspberrypi_redball/2"
SAM_MODEL_NAME = "sam2.1_b.pt"

# Instanciation des mod√®les via les wrappers
detector = RoboflowDetector(model_id=ROBOFLOW_MODEL_ID, api_key=ROBOFLOW_API_KEY)
segmenter = SAMSegmenter(model_name=SAM_MODEL_NAME)

print("‚úì Mod√®les charg√©s avec succ√®s")
```

```python
# Chargement des images √† traiter
imgs_path = Path("df")
assert imgs_path.exists(), f"Dossier '{imgs_path}' non trouv√©"
img_paths = list(imgs_path.glob("*"))

print(f"üìÅ {len(img_paths)} images trouv√©es dans '{imgs_path}'")
```

```python
# Configuration des dossiers de sortie
DET_OUTPUT = Path("detection_output_folder")
SAM_OUTPUT = Path("sam_output_folder")

print(f"üöÄ D√©marrage du traitement de {len(img_paths)} images...")
start_time = time.time()

for img_path in tqdm(img_paths):
    img_pipeline(
        img_path,
        detect_fn=detector.detect,
        segment_fn=segmenter.segment,
        det_output_dir=DET_OUTPUT,
        seg_output_dir=SAM_OUTPUT,
    )

end_time = time.time()
duration = end_time - start_time
print(f"‚úÖ Traitement termin√© en {duration:.2f} secondes.")
print(f"‚è±Ô∏è Moyenne par image : {duration / len(img_paths):.2f} secondes.")
```
